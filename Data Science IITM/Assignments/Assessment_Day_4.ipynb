```python
# Task 3

import json

class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_grade(self):
        average = sum(self.marks) / len(self.marks)
        if average >= 90:
            return "A"
        elif average >= 75:
            return "B"
        elif average >= 60:
            return "C"
        else:
            return "D"

```

```python
student = Student("Narender", 85, [85, 90, 78])
data = {
    "name": student.name,
    "roll_no": student.roll_no,
    "marks": student.marks,
    "grade": student.calculate_grade()
}

with open("student_data.json", "w") as f:
    json.dump(data, f)

print("Student data saved to student_data.json")
```

```python
# Task 4

import os
import shutil

src_folder = "."
dst_folder = "backup"
os.makedirs(dst_folder, exist_ok=True)

for filename in os.listdir(src_folder):
    if filename.endswith(".py"):
        shutil.copy(os.path.join(src_folder, filename), dst_folder)

print("All .py files backed up to 'backup' folder.")
```

```python
def insert_book_details(n):
    for i in range(n):
        book_id = int(input('Book ID: '))
        Name = input('Title: ')       
        Author = input('Author: ')
        Year = int(input('Year: '))

        query = f'''INSERT INTO book (Book_id, Title, Author, Year)
                    VALUES ({book_id}, "{Name}", "{Author}", "{Year}");'''

        cur.execute(query)
        print(f'Book {i+1} inserted successfully.')
```
